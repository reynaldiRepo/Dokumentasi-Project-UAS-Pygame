## Pengertian PyGame

PyGame adalah library cross platform dari python yang diciptakan untuk membaut game. Modul ini dirancang untuk menjadi sederhana, mudah di gunakan dan menyenangkan dngan mengadopsi filosofi dari Python. Sehingga bagi developer yang minim pengalaman bisa langsung menggunakan PyGame secepat mereka mempelajari Python.

## Instalasi PyGame

#### Siapkan file pygame.whl
Download library pygame di https://www.lfd.uci.edu/~gohlke/pythonlibs/ .Sesuaikan versi python yang terinstall dengan versi library pygame yang akan di download. Contoh python v3.5 telah terinstall, maka download library pygame dengan cp35m-win32.

#### Rubah file pygame.whl menjadi .zip
Rubah extension file library pygame lalu extract isi library tersebut.

#### Copy file pygame yang dibutuhkan
1. masuk ke dalam directory python
   (C:\Users\#username\AppData\Local\Programs\Python\Python35-32)
2. masuk kedalam folder “include” dan buat folder baru bernama “pygame”.
3. Di dalam folder hasil extract file library pygame yang sudah didownload, masuk ke “pygame-1.9.4.data\header”, copy semua file di dalam folder tersebut dan masukkan ke dalam folder: C:\Users\#username\AppData\Local\Programs\Python\Python35-32\include\pygame
4. kembali ke folder hasil extract file library pygame tadi, copy folder “pygame” dan “pygame-1.9.4.dist-info” kedalam:
   C:\Users\#username\AppData\Local\Programs\Python\Python35-32\Lib\site-packages
#### Cek hasil installasi
Buka IDLE Python lalu lakukan perintah “import pygame”, jika tidak ada tulisan error maka pygame sudah berhasil terinstall.

### class Segment(pygame.sprite.Sprite):

```python
class Segment(pygame.sprite.Sprite):
    """ Class to represent one segment of the snake. """
    # -- Methods
    # Constructor function
    def __init__(self, x, y, color):
        # Call the parent's constructor
        super().__init__()
        # Set height, width
        self.image = pygame.Surface([segment_width, segment_height])
        self.image.fill(color)
        # Make our top-left corner the passed-in location.
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        
    def check_collision(self, sprite1):
        return pygame.sprite.collide_rect(self, sprite1)
```

Class ini berInherit denga Class pygame.sprite.Sprite class ini mengsung konsep Super.Class untuk melakukan Inheritance dari constructor class ini menyimpan atribut segment yang merupakan Obejek Sprite, dalam class ini juga terdapat method check_collides yang berisikan nilai balikan dari method collides_rect method dari Class Spite.
